```python
#Q9
rows = ['Quiz 1','Quiz 2','Quiz 3','Quiz 4','Quiz 5']
grade['quizaverage']= grade.apply(lambda x:np.mean(x[rows]), axis=1)
grade.head()
len(grade[(grade['quizaverage'] >= 80) & (grade['quizaverage'] < 90)])
```

```python
#10
round(grade.set_index('Major').groupby(level=0)['Quiz 1','Quiz 2','Quiz 3','Quiz 4','Quiz 5','Project 1','Project 2',
              'Team Project','Midterm','Final Exam'].agg(np.average),2)
```

```python
# Q11
grade['droplowestquiz']=grade.apply(lambda x: (np.sum(x[rows])-np.min(x[rows]))/4,axis=1)
grade['finalgrade']=( grade.apply(lambda x: x['droplowestquiz']
                                  *0.12+(x['Project 1']+x['Project 2'])*0.1+
                                  x['Team Project']*0.15+x['Midterm']*0.2
                                 +x['Final Exam']*0.33,axis=1))
np.average(grade['finalgrade'])
```

```python
#Q12 

def checkifexcellent(x):
    if x['finalgrade']>= 92:
        return 1
    else:
        return 0
grade['excellent?']=grade.apply(checkifexcellent,axis=1)
grade.groupby('Major').agg({'excellent?':np.count_nonzero})
```

```python
# Q16
def droptwo(row):
    a= [row['Quiz 1'],row['Quiz 2'],row['Quiz 3'], row['Quiz 4'],row['Quiz 5']]
    a.sort(reverse=True)
    return round((np.sum(a[:3]))/3,2)

grade['drop2quizavg']=grade.apply(droptwo, axis=1)
grade.head()
```
